```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Sample data
data = {
    'Hours': [1, 2, 3, 4, 5],
    'Marks': [35, 50, 60, 70, 85]
}

df = pd.DataFrame(data)
print(df)
```

```python
# Separate input (X) and output (y)
X = df[['Hours']]  # 2D array for features
y = df['Marks']

# Create and train the model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict marks for 6 hours of study
predicted_marks = model.predict([[6]])
print("Predicted marks for 6 hours study:", predicted_marks[0])
```

```python
# Plot original data
plt.scatter(df['Hours'], df['Marks'], color='blue', label='Actual Data')

# Plot regression line
plt.plot(df['Hours'], model.predict(X), color='red', label='Prediction Line')

plt.title("Study Hours vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.legend()
plt.grid(True)
plt.show()
```
